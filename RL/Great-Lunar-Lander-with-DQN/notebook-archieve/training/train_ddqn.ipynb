```python
import gym
# from gym.wrappers import RecordVideo

import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from utils import seed_everything
from collections import deque
from model.ddqn import DoubleDQN
from model.dqn import DQN
```

```python
import wandb
wandb.login()
wandb.init(
    entity="onsen",
    project="ddqn-lunar-lander"
)
```

```python
# Set random seed to 1
seed_everything(1)
```

```python
env = gym.make('LunarLander-v2')

# set seeds
env.seed(1)

# setting up params
lr = 0.0009
epsilon = 1.0
epsilon_decay = 0.96
gamma = 0.99
training_episodes = 1000
target_update_interval = 5
# create new deep q-network instance
model = DoubleDQN(env=env, lr=lr, gamma=gamma, epsilon=epsilon, epsilon_decay=epsilon_decay, target_update_interval=target_update_interval, log_wandb=True)
model.train(training_episodes, True)
```

```python
# Complete training
wandb.finish()
```
